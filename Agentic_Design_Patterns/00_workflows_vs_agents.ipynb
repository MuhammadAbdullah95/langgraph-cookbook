# Agentic Design Patterns with LangGraph — Introduction

Welcome! This series of notebooks walks through the **six core agentic design patterns** as documented by LangChain:

## The augmented LLM — the building block

Every pattern below is built on the same idea: an LLM *augmented* with tools, retrieval, and memory. Agentic systems wire several augmented-LLM calls together.

![An LLM augmented with tools, retrieval, and memory](../images/augmented_llm.avif)

<https://docs.langchain.com/oss/python/langgraph/workflows-agents>

By the end of this class you will understand:

1. **Prompt Chaining** — call LLMs in sequence, each one builds on the last
2. **Parallelization** — run multiple LLM calls at the same time
3. **Routing** — send the request to the right specialist
4. **Orchestrator–Worker** — plan a task, delegate pieces, combine results
5. **Evaluator–Optimizer** — generate, grade, improve, repeat
6. **Agents** — let the LLM decide which tools to use in a loop

---

## Workflows vs Agents

> *"Workflows have predetermined code paths and are designed to operate in a certain order. Agents are dynamic and define their own processes and tool usage."* — LangChain docs

![Overview of workflow and agent patterns](../images/agent_workflow.avif)


Think of it like this:

| | **Workflow** | **Agent** |
|---|---|---|
| Who decides the order? | **You** (the developer) | **The LLM** |
| Path through the code | Fixed | Dynamic — changes every run |
| Best for | Well-understood tasks | Open-ended problems |
| Example | "Translate → grammar check → polish" | "Figure out the user's intent and use any of these 10 tools" |

**Patterns 1–5 are workflows.** You wire the nodes together in advance.
**Pattern 6 is an agent.** The LLM decides which node runs next by calling tools.

Both sit on the same foundation: a **LangGraph `StateGraph`** — nodes are Python functions, edges are the arrows between them, and a shared state dict flows through.

````python
# Setup — run this first in every notebook
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
os.environ["GEMINI_API_KEY"] = os.getenv("GEMINI_API_KEY")

llm = ChatGoogleGenerativeAI(
    model="gemini-3-flash-preview",
    temperature=1.0,
    max_retries=2,
    google_api_key=os.getenv("GEMINI_API_KEY"),
)
````

````python
# Smoke test — this should return a short sentence
reply = llm.invoke("In one sentence, what is LangGraph?")

````
